## Theory

## 1. Fundamentals of Superconductivity



### 1.1 Cooper pair formation (phonon-mediated attraction, BCS theory)

Superconductivity is the phenomenon where the electrical resistance of a metal vanishes below a critical temperature $T_c$. A defining propery is the Meissner effect, the repulsion of magnetic fields from the interior of a superconductor, which shows that superconductivity is a distinct macroscopic quantum state rather than mere perfect conductivity.

London (1937) introduced the idea of a rigid superconducting wavefunction $\psi(\mathbf{r})$. He showed that the current in such a state is
$$ \mathbf{j} = \frac{e\hbar}{2mi}(\psi^* \nabla \psi - \psi \nabla \psi^*) - \frac{e^2}{m}|\psi|^2 \mathbf{A} $$
For a uniform ground state, the paramagnetic term vanishes and one obtains the London equation
$$ \mathbf{j} = -\frac{e^2}{m}|\psi|^2 \mathbf{A} $$
Combining this with Maxwell's equations leads to
$$ \nabla^2 \mathbf{B} = \frac{1}{\lambda^2} \mathbf{B} $$
where $$ \lambda = \sqrt{\frac{m}{\mu_0 e^2 |\psi|^2}} $$
showing that the magnetic fields decay exponentially inside the superconductor with penetration depth $ \lambda $.
While London's description was phenomenological, the microscopic origin of superconductivity was traced to the electron-phonon interaction. THe exchange of a phonon between two electrons leads to an effective attractive interaction,
$$ V_{eff}(\mathbf{q}, ϵ_k ,ϵ_k') \approx - \frac{|g_{\mathbf{q}}|^2}{ \omega_{\mathbf{q}}^2 - (ϵ_k - ϵ_k')} $$
where $ g_{\mathbf{q}} $ is the electron-phonon coupling constant, and $ \omega_{\mathbf{q}} $ is the phonon frequency. This gives an attractive interaction for frequencies below the phonon frequency, $ |ϵ_k-ϵ_k'| < \omega_{\mathbf{q}} $. The isotope effect, where $ T_c ∝ M^{-\frac{1}{2}} $ (M = ion mass), provided strong evidence for the phonon-mediated mechanism.

Building on this, Cooper(1956) considered two electrons above a filled Fermi sea interacting via such an attraction. He showed that an arbitrarily weak interaction destabilizes the Fermi surface, leading to a bound state (Cooper pair) with total momentum zero. THe pair creation operator takes the form
$$ Λ^{†} = \sum_k \phi(k) c_{k↑}^{†} c_{-k↓}^{†} $$
where $ \phi(k) $ is the pair wavefunction. The pair wavefunction satisfies the self-consistent condition
$$ 1 = g_0 N(0) ∫ _0 ω_D  \frac{dϵ}{2ϵ - E} $$
where $ g_0 $ is the interaction strength, $ N(0) $ the density of states at the Fermi level, and $ ω_D $ the Debye frequency. The solution yields the binding energy of a Cooper pair
$$ E ≈ -2ω_D e^{-\frac{1}{g_0 N(0)}} $$
This result is fundamental. No matter how weak the attraction, the Fermi sea is unstable towards the formation of bound electron pairs. These Cooper pairs behave as composite bosons, and their condensation gives rise to the superconducting ground state. This forms the basis of the Bardeen-Cooper-Schrieffer (BCS) theory.


### 1.2 BCS mean-field Hamiltonian, gap equation, BdG equations
Bardeen, Cooper, and Schrieffer (1957) formulated a microscopic theory of superconductivity in terms of a condesate of electron pairs. The essential idea is that the instability of the Fermi surface leads to the formation of a coherent ground state where Cooper pairs condense.

The minimal model capturing the physics of $s$-wave pairing is
$$ H = \sum_{k,σ} ϵ_k c_{kσ}^{†} c_{kσ} + \frac{g_0}{V}  \sum_{k,k'} V_{kk'} c_{k↑}^{†} c_{-k↓}^{†} c_{-k'↓} c_{k'↑} $$

where $\epsilon_k$ is the single-particle energy measured from the Fermi energy, $V$ the volume, and $g_0 > 0$ the effective pairing strength. The interaction acts only within a shell of width $\omega_D$ around the Fermi surface, reflecting its phonon origin.
This Hamiltonian is not exactly solvable in full. The key step is a mean-field decoupling that replaces the quartic interaction by an average over a pairing field.

Introducing the anomalous average
$$ \Delta = -\frac{g_0}{V} \sum_k  \langle c_{-k↓} c_{k↑} \rangle $$
which defines the superconducting gap parameter (complex in general). Applying the mean-field decoupling, 
$$ c_{k↑}^{†} c_{-k↓}^{†} c_{-k'↓} c_{k'↑} \approx \langle c_{k↑}^{†} c_{-k↓}^{†} \rangle c_{-k'↓} c_{k'↑} + c_{k↑}^{†} c_{-k↓}^{†} \langle c_{-k'↓} c_{k'↑} \rangle - \langle c_{k↑}^{†} c_{-k↓}^{†} \rangle \langle c_{-k'↓} c_{k'↑} \rangle $$
The Hamiltonian reduces to the quadratic mean-field form
$$ H_{MF} = \sum_{k,σ} ϵ_k c_{kσ}^{†} c_{kσ} - \sum_k ( \Delta c_{k↑}^{†} c_{-k↓}^{†} + \Delta^* c_{-k↓} c_{k↑} ) + \frac{V}{g_0} |\Delta|^2 $$
This Hamiltonian can be diagonalized exactly.

Bogoliubov transformation introduces new quasiparticle via
$$ \alpha_{k↑} = u_k c_{k↑} - v_k c_{-k↓}^{†} $$
$$ \alpha_{-k↓}^{†} = v_k c_{k↑} + u_k c_{-k↓}^{†} $$
where $u_k$ and $v_k$ are coherence factors satisfying $u_k^2 + v_k^2 = 1$. The mean-field Hamiltonian becomes diagonal in terms of these quasiparticles:
$$ H_{MF} = \sum_k E_k ( \alpha_{k↑}^{†} \alpha_{k↑} + \alpha_{-k↓}^{†} \alpha_{-k↓} ) + \text{const} $$
where the quasiparticle energy spectrum is
$$ E_k = \sqrt{ϵ_k^2 + |\Delta|^2} $$
Thus, the superconducting state has an energy gap $|\Delta|$ protecting it against low-energy excitations.

The self-consistency condition gives the BCS gap equation:
$$ 1 = \frac{g_0}{V} ∑_{|ϵ_k| < ω_D}^{} \frac{1}{2E_k} \tanh\left(\frac{E_k}{2k_B T}\right) $$
At zero temperature, this reduces to
$$ 1 = g_0 N(0) \ln\left(\frac{2ω_D}{|\Delta|}\right) $$
yielding the zero-temperature gap
$$ |\Delta(0)| = 2ω_D e^{-\frac{1}{g_0 N(0)}} $$
in agreement with Cooper's result.
The universal BCS relation follows:
$$ 2|\Delta(0)| \approx 3.53 k_B T_c $$

#### The Bogoliubov-de Gennes (BdG) equations
The mean-field Hamiltonian can also be expressed in Nambu spinor form:
$$ \Psi_k = \begin{pmatrix} c_{k↑} \\ c_{-k↓}^{†} \end{pmatrix} $$
The Hamiltonian becomes
$$ H_{MF} = ∑_{k}^{} Ψ_k^{†} \begin{pmatrix} ϵ_k & \Delta \\ \Delta^* & -ϵ_k \end{pmatrix} Ψ_k$$
This $2 \times 2$ matrix form generalizes to spatially inhomogeneous systems, yielding the bogoliubov-de Gennes (BdG) equations:
$$ \begin{pmatrix} H_0(\mathbf{r}) & \Delta(\mathbf{r}) \\ \Delta^*(\mathbf{r}) & -H_0^*(\mathbf{r}) \end{pmatrix} \begin{pmatrix} u_n(\mathbf{r}) \\ v_n(\mathbf{r}) \end{pmatrix} = E_n \begin{pmatrix} u_n(\mathbf{r}) \\ v_n(\mathbf{r}) \end{pmatrix} $$
where $H_0(\mathbf{r})$ is the single-particle Hamiltonian including kinetic and potential terms. The BdG equations describe quasiparticle excitations in spatially varying superconducting systems, such as near interfaces or impurities.

### 1.3 Ginzburg-Landau theory (qualitative)
Ginzburg-Landau (GL) theory (1950) provides a phenomenological description of superconductivity, formulated before the microscopic BCS framework. It extends Landau's theory of second order phase transitions by introducing a complex order parameter 
$$ \psi(\mathbf{r}) = |\psi(\mathbf{r})| e^{i\theta(\mathbf{r})} $$
whose magnitude $|\psi|^2$ represents the local density of superconducting electrons.

The free energy density of a superconductor near the critical temperature $T_c$ is expanded as
$$ f_s = f_n + \alpha(T)|ψ|² + \frac{1}{2} \beta(T)|ψ|^4 + \frac{1}{2m^*} |(-i\hbar\nabla - \frac{e*}{c}\mathbf{A})ψ|² + \frac{|\mathbf{B}|²}{8 π} $$
where $f_n$ is the normal state free energy, $\alpha(T)$ and $\beta(T)$ are phenomenological coefficients, $m^*$ and $e^*$ are the effective mass and charge of the Cooper pairs, $\mathbf{A}$ is the magnetic vector potential, and $\mathbf{B} = \nabla \times \mathbf{A}$ is the magnetic field.
Minimizing this functional yields the Ginzburg-Landau equations for the order parameter and current, showing that below $T_c$, a nonzero $\psi$ develops continuously, typical of a second-order phase transition.

GL theory introduces two characteristic length scales:
1. The coherence length $\xi(T)$, which describes the spatial variation of the order parameter ($ψ$)
$$ \xi(T) = \sqrt{\frac{\hbar²}{2m^* |\alpha(T)|}} $$
2. The penetration depth $\lambda(T)$, which characterizes the decay of magnetic fields inside the superconductor
$$ \lambda(T) = \sqrt{\frac{m^* c²}{4 π e*² |\psi|²}} $$
The ratio $\kappa = \frac{\lambda}{\xi}$ classifies superconductors into two types:
- Type I: $\kappa < \frac{1}{\sqrt{2}}$, where the magnetic field is completely expelled (Meissner effect).
- Type II: $\kappa > \frac{1}{\sqrt{2}}$, where magnetic vortices can penetrate the material.

Though phenomenological Gor'kov later derived GL theory from BCS in the limit $T ≲ T_c$, giving macroscopic meaning to its parameters. GL thus bridges macroscopic superconductivity (phase transitions, vortices) and microscopic pairing physics, and remains a central tool in analyzing superconductors with spatial inhomogeneities.

## 2. Interfaces and Hybrid Systems



### 2.1 Proximity effect: placing a SC on a SM (mathematics of induced pairing)
The proximity effect is a phenomenon where a non-superconducting material, such as a semiconductor (SM), aquires superconducting properties when placed in contact with a superconductor (SC). This is a cornerstone of creating hybrid systems designed to engineer novel quantum states, like topological superconductivity, from conventional materials. This strategy is often called a "top-down" approach, as it builds complex states from well-understood components.

The core mechanisms involves the leakage of Cooper pairs from the SC into the SM. This generates spin-singlet Cooper pairs within the semiconductor, resulting in an induced superconducting gap, denoted as $\Delta_{ind}$. The SM, therefore, effectively becomes superconducting, a crucial step for engineering more complex states.

To create a topological phase in a semiconductor nanowire, three competing effects must be balanced:
1. **S-wave Superconducting Proximity Effect**: This induces the gap $\Delta_{ind}$ in the SM, establishing the necessary superconducting correlations.
2. **Zeeman Energy**: An external magnetic field, B, is applied to break spin degeneracy. The Zeeman energy, $E_Z = \frac{1}{2} g \mu_B B$, tends to break the spin-singlet Cooper pairs by aligning electron spins, which works to close the induced gap.
3. **Spin-Orbit Coupling (SOC)**: This intrinsic property of the SM counteracts the Zeeman field by preventing the electron spins from achieving full alignment. This allows the induced superconductivity to survive even under a strong magnetic field.

The competition between these effects can drive a topological phase transition. This transition - the closing and reopening of the induced gap - occurs when the Zeeman energy matches the energy cost associated with the chemical potential and induced pairing. The mathematical condition for this topological transition is given by:
$$ E_Z = \sqrt{\Delta_{ind}^2 + \mu^2} $$
where $\mu$ is the chemical potential in the SM. Above this critical Zeeman field, the hybrid system effectively behaves as a topological spinless p-wave superconductor, akin to the Kitaev chain model.

For this to be experimentally viable, certain conditions are desirable:
- A large g-factor in the semiconductor relative to the superconductor (|g/g_SC| >> 1) to reach the topological regime without destroying the parent superconductivity.
- A low carrier concentration (small $\mu$) in the semiconductor, which minimizes the magnetic field required to induce the transition.
- A large induced gap $\Delta_{ind}$ and strong SOC are needed to ensure the resulting Majorana states are well-protected and separated from other states.

### 2.2 Andreev reflection and Cooper pair splitting
Andreev reflection (AR) is a fundamental charge-transfer process that occurs at the interface between a normal conductor (N) and a superconductor (SC). When an electron is incident on the interface from the normal side with an energy less than the superconducting gap $\Delta$, it cannot enter the SC as a single quasiparticle. Instead, it is retroreflected as a hole, and a Cooper pair is coherently transferred into the superconducting condensate. 

A non-local version of this process, crucial for hybrid systems, is Crossed Andreev Reflection (CAR), also known as Cooper pair splitting. In CAR, a Cooper pair from a central superconductor splits, and its two constituent electrons tunnel into two spatially separated terminals, such as two different quantum dots (QDs). For this to occur, the distance between the two terminals must be less than the superconducting coherence length $\xi$.

The typical setup for studying CAR is a double quantum dot (DQD) system coupled by a central superconductor (QD-SC-QD). In such a device, two distinct processes compete:
- ** Local Andreev Reflection (LAR)**: A Cooper pair tunnels from the SC into a single QD. This process is strongly suprecced by the QD's charging energy $U$ (Coulomb blockade), which penalizes the addition of two electrons at once. This supreccsion is a key advantage of using QDs.
- ** Crossed Andreev Reflection (CAR)**: An electron tunnels into each of the two QDs, a process that is not suppressed by local charge energy and is therefore the dominant Andreev process in this geometry.

The role of CAR is fundamental in the bottom-up engineering of topological systems. In the context of creating a minimal Kitaev chain with two QDs, CAR provides the non-local pairing term that is mathematically equivalent to the p-wave pairing amplitude ($Δ$) in the Kitaev model. It acts in concert with Elastic Co-Tunneling (ECT), where an electron hops between the dots via the superconductor, which provides the nearest-neighbor hopping term ($t$) in the Kitaev chain.

The magnitude of the CAR process can be characterized by measuring the current flowing from the superconductor into the leads attached to the QDs. Under resonant conditions ($ϵ_L = - ϵ_R$) and specific biasing, the current due to CAR is given by:
$$ I_{CAR} = \frac{e}{\bar{h}} \frac{Γ}{(ϵ_L + ϵ_R)^2 + Γ²} |Δ|²$$
where $Γ$ is the coupling to the leads, and $Δ$ represents the effective CAR amplitude. By measuring the current for both CAR and ECT, it is possible to tune the system to a sweet spot where $|t| = |Δ|$, a condition necessary for forming Majorana bound states (MBS). The primary application of Cooper pair splitting is the creation of entangled electron pairs for quantum information and the bottom-up construction of artificial topological superconductors.


### 2.3 Formation of ABS
Andreev Bound States (ABS) are discrete, localized quasiparticle states with energies that lie inside the superconducting gap ($Δ > E$). They are a hallmark of hybrid superconductor systems and form whenever electron and hole quasiparticles are coherently confined to a specific region. This confinement can be caused by various mechanisms, such as interfaces, impurities, or weak links between superconductors.

The formation of an ABS can be understood as the trapping of a quasiparticle between two Andreev reflection events. These states are not pure electron or hole states but rather a coherent superposition of both, described by the Bogoliubov-de Gennes (BdG) coefficients $u$ and $v$. The relative weight of these coefficients ($|u|^2$ and $|v|^2$) determines the "BCS charge" of the state, indicating whether it is more electron-like or hole-like.

Two primary scenarios for the formation of ABS is semiconductor-superconductor (SM-SC) devices.

- **Geometric Confinement**: When a segment of a semiconductor, like quantum dot, is coupled to a superconductor, it acts as a potential well that confines quasiparticles. An ABS forms at the interface, with an energy that is highly tunable. The energy of such a state can be described by the expression:
$$E = \sqrt{(ϵ-μ)²-Δ_0^²}$$
where $ϵ$  is the energy of the confined state in the QD, $μ$ is the chemical potential, and $Δ_0$ is the superconducting gap. The energy of this ABS is minimized when the chemical potential aligns with the dot level ($μ = ϵ$).

- **Magnetic Impurities (Yu-Shiba-Rusinov States)**: When a QD coupled to a superconductor is tuned to have an odd number of electrons, its net spin acts as a local magnetic impurity. This leads to the formation of a specific type of ABS known as a Yu-Shiba-Rusinov (YSR) state. The exchange interaction between the local spin on the QD and the Cooper pairs in the superconductor binds the subgap state to the impurity site. 

A single spin-degenerate ABS can be thought of as a four-level quantum system. It has two states with even fermion parity (the ground state with no quasiparticles $\ket{g}$, and an excited state with two quasiparticles in the ABS $\ket{e}$) and two states with odd fermion parity, corresponding to a single quasiparticle with either spin-up or spin-down. These states are fundamental building blocks for various quantum devices, including Andreev qubits and as mediators for processes like CAR and ECT between other quantum dots.

## 3. Josephson Junctions


### 3.1 Current-Phase Relation
The Current-Phase Relation (CPR) is the fundamental equation for the Josephson effect, describing the relationship between the supercurrent $I_s$, that flows through a Josephson-junction (JJ) and the gauge-invariant quantum mechanical phase difference $\varphi$ between the two superconducing electrodes.

#### Standard (Sinusoidal) CPR
For a conventional junction with a simple insulating barrier (a SIS junction), the CPR takes a simple sinusoidal form, often referred to as the first Josephson equation:

$$ I_s = I_c \sin(\varphi) $$

where $I_c$ is the critical current of the junction, which represents the maximum supercurrent that can flow through the junction without generating voltage drop (meaning without resistance). This relationship is based on the general principle that the supercurrent must be a $2\pi$ periodic function of the phase and must be zero when the phase difference is zero.

#### General and Anomalous CPRs
In many realistic scenarios, especially in junctions with high barrier transparancy or those incorporating materials with complex properties (like ferromagnets), the CPR deviates from simple sinusoidal. The standard CPR is the first term in a more general Fourier series expansion:
$$I_s (\varphi) = ∑_{n\geq 1}^{}I_n \sin(n \varphi) $$
The higher harmonic terms ($I_n$ for $n \geq 2$) become significant as the barrier transparency increases and are associated with multiple Andreev reflections processes.

Anomalous CPRs arise from symmetry breaking effects, which are deviations from the standard CPR. They often arise from the breaking of fundamental symmetries:
- **$π$-junctions**: In certain configurations, such as superconductor-ferromagnet-superconductor (SFS) junctions, the ground state of the junction can have an instrinsic phase shift of $π$. This corresponds to an effective negative critical current ($I_c < 0$), and the CPR is dominated by a $-\sin(\varphi)$ term. The Josephson energy is minimized at $φ = π$ instead of $φ = 0$.
- **$\phi_0$-Junctions**: If both time-reversal symmetry (TRS) and inversion symmetry are broken, the junction cane xhibit a spontaneous, non-trivial phase shift $\varphi_0$, in its ground state. The CPR for such a $ϕ_0$-junction is: $I_s(\varphi)=I_c\sin(\varphi-\varphi_0)$. A key feature of a $ϕ_0$-junction is the presence of an anomalous Josephson current - a supercurrent that flows even at zero phase difference: $I_s(0) = -I_c\sin(\varphi_0)$. These junctions are crucial for developing devices like josephson phase batteries.

### 3.2 Role in topological systems (Majorana coupling across junctions)
JJs play a dual role in the pursuit of topological quantum computation. THey can be used not only to induce a topological phase in a material but also to mediate the coupling between MBS. This coupling fundamentally alters the physics of the junction, leading to unique, measurable signatures.

A planar JJ, formed by placing two superconductors on top of a semiconductor, can be used to engineer a topological phase. In a highly transparent junction, the energy gap that protects the superconducting state naturally close as the phase difference across the junction approaches $π$. In this near-gapless regime, even a small external Zeeman field can be sufficient to drive the system through a topological phaset ransition, creating MBS at the boundries of the junction. This makes the JJ an exceptionally tunable platform for creating topological superconductivity.

When MBS are present on either side of a junction, they can couple to each other through the junction. This interaction modifies the standard CPR and is a key feature of topologicals superconductivity.
A prime example is the Majorana Transmon Qubit, an extension of the standard transmon qubit where the Josephson coupling is mediated by two topological superconducting wires. The Hamiltonian for such a system illustrates the interplay between conventional and topological effects:
$$H = H_C+H_J+H_M$$
Here
- $H_C=4E_C(n-n_g)²$ is the standard charging energy of the superconducting island.
- $H_J=E_J\cos(\varphi)$ is the conventional Josephson energy arising from tunneling of Cooper pairs across the junction.
- $H_M=iE_mγ_2γ_3$ is the new energy term describing the coupling between the "inner" Majorana states ($γ_2$ and $γ_3$) located on the opposite sides of the junction.

This Majorana-mediatied coupling facilitates the tunneling of a single electron across the junction, a process that is proportional to $\cos(\frac{\varphi}{2})$. This leads to an energy phase relation that is $4π$-periodic, in stark contrast to the standard $2π$-periodicity associated with Cooper pair tunneling. This phenomenon is known as the fractional Josephson effect and is a hallmark of Majorana coupling. Experimentally, it manifests as the suppression of certain energy level transitions at a phase difference of $\varphi=π$ in the microwave spectrum of the device.

## 4. Topological Superconductivity

### 4.1 Kitaev Chains
The Kitaev Chain is the quintessential "toy model" for understanding topological superconductivity. It describes a one-dimensional tight-binding chain of spinless fermions that can exhibit a topological phase hosting Majorana bound states at its ends. Despite its simplicity, it contains all the essential ingredients for topological superconductivity: spinless fermions (implying broken time-reversal symmetry) and unconventional p-wave superconducting pairing, where electrons on adjecent sites are paired.

The Hamiltonian for the Kitaev chain is given by:
$$ H = -μ ∑_{j=1}^{N} (c_j^\dagger c_j - \frac{1}{2}) + ∑_{j=1}^{N-1} \big[-t (c_j^\dagger c_{j+1} + c_{j+1}^\dagger c_j) + Δ (c_j c_{j+1} + c_{j+1}^† c_j )   \big]$$
where $μ$ is the onsite energy, $t$ is the hopping amplitude, and $Δ$ is the p-wave pairing amplitude. 

#### 4.1.1 Band structure, topological invariants
The properties of the Kitaev chain are defined by its bulk energy spectrum (or band structure), which can be found by transforming the Hamiltonian into momentum space. The resulting excitation spectrum for the quasiparticle is:
$$E_{p \pm} = \pm \sqrt{(μ - 2t \cos(pa))² + 4Δ^2 \sin^2(pa)}$$
where $p$ is the momentum and $a$ is the lattice spacing.

The spectrum is gapped everywhere except when the term inside the square root is zero. This gap closing occurs when $|μ|=|2t|$, which marks the boundry between two distinct phases of matter: a trivial phase and a topological phase.

These phases are distinguished by a $\mathbb{Z}_2$ topological invariant known as the Majorana number, $M=(-1)^v$, where $v$ is the number of pairs of Fermi points.
- **Trivial phase**: Occurs for $|μ|>|2t|$.In this regime, the number of Fermi point pairs is even ($v$ is even), so the Majorana number $M=1$.
- **Topological phase**: Occurs for $|μ|<|2t|$. Here, the number of Fermi point pairs is odd ($v$ is odd), leading to a Majorana number $M=-1$.
The existance of a non-trivial topological invariant ($M=-1$) in the bulk of the material dictates the presence of protected states at the boundries, a principle known as the bulk-boundry correspondence.

#### 4.1.2 Majorana edge states in finite chains
In a finite chain, the distinction between the topological and trivial phases becomes physically apparent through the presence or absence of zero-energy states at the ends of the chain. This can be understood by decomposing each fermion operator $c_j$ into two real Majorana operators, $γ_j^A$ and $γ_j^B$:

$$c_j = \frac{1}{2}(γ_j^A + iγ_j^B)$$
$$c_j^{†} = \frac{1}{2}(γ_j^A - iγ_j^B)$$
The two phases correspond to different ways of pairing Majorana operators:
-**Trivial Phase**: When the on-site energy $μ$ dominates ($t=Δ=$), the Majorana operators on the same site ($γ_j^A$ and $γ_j^B$) couple together to form a standard, localized fermion. There are no unpaired modes at the ends.
-**Topological Phase**: In the topological regime (at the "sweet spot" where $μ=0$ and $t=Δ$), the Hamiltonian simplifies such that Majorana operators on neighboring sites couple together ($γ_j^B$ couples with $γ_{j+1}^A$). This leaves the very first Majorana at the start of the chain, $γ_1^A$, and the very last Majorana at the end of the chain, $γ_N^B$, completely unpaired.

These two unpaired, spatially separated Majorana operators are the Majorana edge states, also known as Majorana zero modes (MZMs) or Majorana bound states (MBSs). They can be combined to form a single, non-local fermionic state $f$ that has zero energy:

$$f = \frac{1}{2}(γ_1^A + iγ_N^B)$$
Because this state has zero energy, the ground state of the topological chain is twofold degenerate, corresponding to this non-local fermion state being empty or occupied. This degeneracy is topologically protected, meaning the edge states cannot be removed or given a finite energy by any small, local perturbation without closing the bulk energy gap.

## 5. Experimental Realizations

### 5.1 Nanowires
The leading platform for realizing one-dimensional topological superconductivity is the semiconductor-superconductor (SM-SC) nanowire. The goal is to engineer the conditions within the nanowire to create an "effective" p-wave superconductor that hosts Majorana bound states (MBS) at its ends, thereby realizing a physical version of the Kitaev model.



#### 5.1.1 Ingredients: strong spin-orbit coupling, Zeeman field, induced superconductivity
The creation of a topological phase in a hybrid nanowire relies on the precise tuning and competition between three key physical effects:
1. **Induced Superconductivity**: A conventional s-wave superconductor (like aluminium) is placed in close proximity to a semiconductor nanowire (like InAs or InSb). Through the proximity effect, Cooper pair correlations leak into the nanowire, inducing a superconducting gap ($\Delta_{ind}$).
2. **Zeeman Field**: A strong external magnetic field ($B$) is applied, typically parallel to the nanowire. This creates a large Zeeman energy ($E_Z = \frac{1}{2} g \mu_B B$) that lifts the spin degeneracy of the electrons in the nanowire. This effects tends to break the spin-singlet Cooper pairs, which works to close the induced gap.
3. **Spin-Orbit Coupling (SOC)**: The semiconductor material is chosen for its intrinsically strong SOC. This coupling prevents the electron spins from fully aligning with the magnetic field, allowing the induced superconductivity to persist even at high Zeeman energies.

The competition between these ingredients drives a topological phase transition. When the Zeeman energy is strong enough to close and reopen the gap (specifically, when $E_Z > \sqrt{μ² + \Delta_{ind}}$), the system enters a topological phase that is predicted to host MBS. 


#### 5.1.2 Zero-bias peaks and limitations
The primary experimental signature for the presence of MBS in nanowires is a zero-bias conductance peak (ZBCP). In a tunneling spectroscopy measurement, where a voltage is applied between a normal lead and the end of a nanowire, a sharp peak in conductance at zero voltage indicates the emergence of a new, zero-energy state available for tunneling, which is consistent with the presence of Majorana Modes.

However, a ZBCP is not conclusive proof of topological MBS, and this ambiguity is the central limitation in the field. There are several non-topological phenomena that can mimic this signature:
- **Trivial ABS**: In the presence of disorder or a smooth confinement potential within the nanowire, conventional ABS can be accidentally tuned to zero energy, producing a ZBCP that is difficult to distinguish from a true MBS. These are sometimes reffered to as "quasi-Majoranas" or "pseudo-Majoranas".
- **Unintentional Quantum Dots**: Imperfections in the nanowire can create localized states that behave like quantum dots. These dots can host Yu-Shiba_Rusinov (YSR) states, which can also cross zero energy and produce a ZBCP as system parameters are tuned.

Because these trivial zero-energy states can mimic the local properties of MBS, a definitive demonstration of topological superconductivity requires probing the non-local nature of the states, for instance through braiding experiments.

### 5.2 QD-SC-QD (PMMs)
An alternative, "bottom-up" approach to realizing a Kitaev chain involves using discrete quantum dots (QDs) coupled via a superconductor (SC). This QD-SC-QD setup creates a "minimal Kitaev chain" that can host Poor Man's Majoranas (PMMs). Majorana states that share the properties of their topological counterparts but lack the full topological protection of a long wire. This platform offers high tunability and control over the system's parameters. 

#### 5.2.1 How proximity + Coulomb effects can create effective Kitaev chains
The QD-SC-QD system engineers an effective Kitaev chain by mimicking its essential physical processes through proximity-induced effects. The two spin-polarized QDs acts as the two sites of a minimal chain, and the central superconductor mediates the necessary couplings:
- **Effective Hopping ($t$)**: An electron can tunnel from one QD to the other through a virtual state in the superconductor. This process, known as Elastic Cotunneling (ECT), serves as the nearest-neighbor hopping term $t$ in the Kitaev model.
- **Effective p-wave Pairing ($Δ$)**: A Cooper pair from the superconductor can split, with each electron tunneling into a different QD. This Crossed Andreev Reflection (CAR) process provides the non-local pairing term $Δ$ required by the Kitaev model.

For Majoranas to form the system must be tuned to a "sweet spot" where the amplitudes of these two processes are equal ($|t| = |Δ|$) and the QD energy levels are aligned with the superconductor's chemical potential.

Coulomb effects play a crucial role. Local charging energy within each QD is beneficial, as it suppresses unwanted Local Andreev Reflection (LAR) and helps polarize the electron spins in a magnetic field. However, non-local Coulomb interaction between the dots is detrimental as it can spoil the conditions needed for well-localized Majoranas. Fortunately, this inter-dot interaction is typically screened by the central superconductor.

#### 5.2.2 Theoretical model, Hamiltonian, simulation of low-energy modes
The minimal theoretical model for a PMM system, assuming fully spin-polarized QDs, is given by the Hamiltonian:
$$ H = \sum_{i=L,R} ϵ_i d_i^{†} d_i + t (d_L^{†} d_R + d_R^{†} d_L) + Δ (d_L d_R + d_R^{†} d_L^{†}) $$
where $d_i^{†}$ ($d_i$) creates (annihilates) an electron in QD $i$ with energy level $ϵ_i$, $t$ is the ECT amplitude, and $Δ$ is the CAR amplitude. At the sweet spot ($ϵ_L = ϵ_R = 0$ and $|t| = |Δ|$), the system hosts two zero- eigenstates which are Majorana modes perfectly localized on each dot.

More realistic microscopic models account for both spin species, local and non-local Coulomb interactions ($U$), and the properties of the media properties of the mediating superconductor. The quality of the resulting Majorana states is quantified by the Majorana Polarization (MP), a value that describes how closely the local wavefunction on a QD resembles an ideal Majorana. Simulations of these more complex models show that high-quality sweet spots (degenerate ground states, high MP in both dots, and a substantial gap to excited states) can still be found by tuning the gate voltages that control the energies of the QDs and the central superconducting region.

#### 5.2.3 Potential for braiding and gate operations
PMMs, despite their lack of full topological protection, are predicted to exhibit the same non-Abelian statistics as their topological counterparts, making them a promising platform for demonstrating braiding operations. Because the ground state degeneracy is only present at the fine-tuned sweet spot, the system can be easily initialized into a well-defined state by simply detuning the dot energies to lift the degeneracy. The final state cna be read out using parity-to-charge conversions, where detuning the dots from the sweet spot makes the charge distribution dependent on the system's fermion parity.

Several protocols, adapted from proposals for topological wires, have been proposed to demonstrate non-Abelian properties with PMMs:
- **Charge-transfer Based Braiding**: Uses an ancillary QD to shuttle single electrons onto the PMM system. The final state of the system depends on the sequence of these charge transfers, which can be made equivalent to a braiding operation.
- **Hybridization-Based Braiding**: In a system with at least three PMMs, the coupling (hybridization) between them can be dynamically switched on and off in a specific sequence to effectively move the majorana states and perform a braiding operation.
- **Measurement-Based Braiding**: The joint fermion parity of PMMs is projectively measured in a specific sequence. A successful sequence of post-selected outcomes is mathematically equivalent to braiding.
